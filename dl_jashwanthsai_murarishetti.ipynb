```python
import gdown
import pandas as pd

# Download the datasets from Google Drive
fear_greed_id = '1PgQC0tO8XN-wqkNyghWc_-mnrYv_nhSf'
historical_data_id = '1IAfLZwu6rJzyWKgBToqwSmmVYU6VbjVs'

gdown.download(f'https://drive.google.com/uc?id={fear_greed_id}', 'fear_greed_index.csv', quiet=False)
gdown.download(f'https://drive.google.com/uc?id={historical_data_id}', 'historical_trader_data.csv', quiet=False)

# Load the datasets into pandas DataFrames
fear_greed_df = pd.read_csv('fear_greed_index.csv')
historical_data_df = pd.read_csv('historical_trader_data.csv')

# Display the first few rows of each DataFrame
print("Fear Greed Index DataFrame:")
display(fear_greed_df.head())

print("\nHistorical Trader Data DataFrame:")
display(historical_data_df.head())
```

```python
# 6. Convert the 'Timestamp IST' column in historical_data_df to datetime objects.
# Explicitly providing the format based on the error message: "18-03-2025 12:50" seems like day-month-year
historical_data_df['Timestamp IST'] = pd.to_datetime(historical_data_df['Timestamp IST'], format='%d-%m-%Y %H:%M')
print("\nData types of historical_data_df after converting 'Timestamp IST' column:")
display(historical_data_df.info())
```

```python
historical_data_df['date'] = historical_data_df['Timestamp IST'].dt.date
display(historical_data_df.head())
```

```python
historical_data_df['date'] = pd.to_datetime(historical_data_df['date'])
merged_df = pd.merge(historical_data_df, fear_greed_df, on='date', how='inner')
display(merged_df.head())
```

```python
performance_by_sentiment = merged_df.groupby('classification')['Closed PnL'].agg(['mean', 'median'])
print("\nTrader performance metrics (Mean and Median Closed PnL) by Market Sentiment Classification:")
display(performance_by_sentiment)
```

```python
import matplotlib.pyplot as plt

# Create a bar chart for mean Closed PnL
plt.figure(figsize=(10, 6))
plt.bar(performance_by_sentiment.index, performance_by_sentiment['mean'], color='skyblue')
plt.xlabel('Market Sentiment Classification')
plt.ylabel('Mean Closed PnL')
plt.title('Mean Trader Performance by Market Sentiment')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Create a bar chart for median Closed PnL
plt.figure(figsize=(10, 6))
plt.bar(performance_by_sentiment.index, performance_by_sentiment['median'], color='lightcoral')
plt.xlabel('Market Sentiment Classification')
plt.ylabel('Median Closed PnL')
plt.title('Median Trader Performance by Market Sentiment')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
